```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'F:\AI\Super AI SS4\Level 3 - INTERN\Jupyter Notebook\Latest-Dataset-Model-Generate\Edit-Prompt\Gemma-latest-prompt-score.csv')
df2 = pd.read_csv(r'F:\AI\Super AI SS4\Level 3 - INTERN\Jupyter Notebook\Latest-Dataset-Model-Generate\Edit-Prompt\Llama3-latest-prompt-score.csv')
df3 = pd.read_csv(r'F:\AI\Super AI SS4\Level 3 - INTERN\Jupyter Notebook\Latest-Dataset-Model-Generate\Edit-Prompt\Qwen2-latest-prompt-score.csv')
```

```python
index_to_drop = [3, 4, 12, 13, 16]

df1 = df1.drop(index=index_to_drop)
df2 = df2.drop(index=index_to_drop)
df3 = df3.drop(index=index_to_drop)
```

```python
df2
```

```python
def remove_before_colon(text):
    if ':' in text:
        return text.split(':', 1)[1].strip()
    return text

df2.loc[14, 'sum_extractive'] = remove_before_colon(df2.loc[14, 'sum_extractive'])
```

```python
import matplotlib.pyplot as plt
```

```python
average_df1 = df1.drop(columns=['index_original']).select_dtypes(include=['number']).mean()
average_df2 = df2.drop(columns=['index_original']).select_dtypes(include=['number']).mean()
average_df3 = df3.drop(columns=['index_original']).select_dtypes(include=['number']).mean()

comparison_df = pd.DataFrame({
    'Gemma2': average_df1,
    'Llama3': average_df2,
    'Qwen2': average_df3
})

comparison_df = comparison_df.dropna()

plt.figure(figsize=(10, 6))
comparison_df.plot(kind='bar', color=['#FF9999', '#66B3FF', '#99FF99'])
plt.title('Average Scores Comparison')
plt.ylabel('Average Score')
plt.xlabel('Scores')
plt.xticks(rotation=45)
plt.legend(title='Models')
plt.tight_layout()
plt.show()
```

```python
average_comparison = pd.DataFrame({
    'Gemma2': average_df1,
    'Llama3': average_df2,
    'Qwen2': average_df3
})

# สร้างกราฟแท่งสำหรับ BERT_precision
plt.figure(figsize=(8, 5))
plt.bar(average_comparison.columns, average_comparison.loc['BERT_precision'], color=['#FF9999', '#66B3FF', '#99FF99'])
plt.title('Average BERT Precision Comparison')
plt.ylabel('Average BERT Precision Score')
plt.xlabel('Models')
plt.xticks(rotation=45)
plt.ylim(0, average_comparison.loc['BERT_precision'].max() + 0.1)
plt.tight_layout()
plt.show()

print("Average values for Gemma2:")
print(average_df1)
print("\nAverage values for Llama3:")
print(average_df2)
print("\nAverage values for Qwen2:")
print(average_df3)
```
