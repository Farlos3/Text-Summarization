```python
def summarize_extractive(text, llm):
    system_template = """You are an AI specialized in extractive summarization for economic news articles."""

    prompt = """Your task is to summarize the key content from the given news article. Follow these guidelines:
    1. Summarize the content in Thai.
    2. Use \t at the beginning of each paragraph to create indentation.
    3. DO NOT leave blank lines between paragraphs. All paragraphs must be continuous with no blank lines.
    4. Focus on main points and important secondary points.
    5. Provide explanations of the article’s key topics without going into too much detail.
    6. Preserve all proper nouns such as names of people, companies, or organizations.
    7. Use 2-3 key sentences from the original article for each point.
    8. Maintain the original meaning and context.
    9. The summary length should not exceed 50% of the original content.
    10. Arrange the content in the same order as presented in the original article.
    11. Reduce redundancy by combining similar points or information.
    12. DO NOT include any examples in the summary.

    IMPORTANT: 
    - Pay close attention to the length of your summary. If the resulting summary is too long, it is crucial to cut out less essential parts. The final summary MUST NOT exceed 50% of the original content's length.
    - DO NOT include any examples or case studies in the summary. Focus only on the main points and key information.

    Article to summarize:
    {text}
    """

    system_message_prompt = SystemMessagePromptTemplate.from_template(system_template)
    human_message_prompt = HumanMessagePromptTemplate.from_template(prompt)

    chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
    messages = chat_prompt.format_messages(text=text)

    return llm.invoke(messages).content
```

#### Repeat summarize

```python
import pandas as pd
from attacut import tokenize
from tqdm import tqdm
import re

# count-word
def count_thai_words(text):
    text = re.sub(r'\s+', ' ', text).strip()
    words = tokenize(text)
    return len(words)

# percent of centent
def calculate_summary_percentage(original_text, summary):
    original_length = count_thai_words(original_text)
    summary_length = count_thai_words(summary)
    percentage = (summary_length / original_length) * 100
    return percentage
```

```python
def summarize_extractive_2(text, llm):
    system_template = """You are an AI specialized in extractive summarization for economic news articles."""

    prompt = """Your task is to summarize the key content from the summary you provided earlier. Follow these guidelines:
    1. Summarize the content in Thai.
    2. Use \t at the beginning of each paragraph to create indentation.
    3. DO NOT leave blank lines between paragraphs. All paragraphs must be continuous with no blank lines.
    4. Focus on main points and important secondary points without going into excessive detail.
    5. Preserve all proper nouns such as names of people, companies, or organizations.
    6. Use 2-3 key sentences from the previous summary for each point.
    7. Maintain the original meaning and context as much as possible.
    8. Include important numbers or statistics in the summary.
    9. The new summary length should be between 30% and 50% of the original content's length.
    10. Arrange the content in the same order as presented in the previous summary.
    11. Reduce redundancy by combining similar points or information.
    12. DO NOT include any examples or illustrations in the summary.

    IMPORTANT: 
    - The final summary MUST be shorter than the previous one but should not exceed 50% of the original content's length.
    - Focus only on the main points and key information from the previous summary.

    Previous summary:
    {text}
    """

    system_message_prompt = SystemMessagePromptTemplate.from_template(system_template)
    human_message_prompt = HumanMessagePromptTemplate.from_template(prompt)

    chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])
    messages = chat_prompt.format_messages(text=text)

    return llm.invoke(messages).content
```

```python
def process_second_round_summarization(df, llm):
    for idx, row in tqdm(df.iterrows(), total=df.shape[0]):
        original_content = row['content']
        first_summary = row['sum_extractive']
        
        summary_percentage = calculate_summary_percentage(original_content, first_summary)

        if summary_percentage > 55:
            second_round_summary = summarize_extractive_2(first_summary, llm)
            
            second_summary_percentage = calculate_summary_percentage(original_content, second_round_summary)
            second_summary_percentage_first = calculate_summary_percentage(first_summary, second_round_summary)
            
            if second_summary_percentage_first > 30 and second_summary_percentage < 55:
                df.at[idx, 'sum_extractive'] = second_round_summary

    return df
```

```python
df2 = process_second_round_summarization(df, llm)
```

```python
def calculate_percentage_for_df(df):
    percent_list = []
    for idx, row in df.iterrows():
        original_content = row['content']
        first_summary = row['sum_extractive']
        
        summary_percentage = calculate_summary_percentage(original_content, first_summary)
        percent_list.append((idx, summary_percentage))
        
    return percent_list

percent_list = calculate_percentage_for_df(df)

for idx, percentage in percent_list:
    print(f"Row {idx}: {percentage:.2f}%")
```
